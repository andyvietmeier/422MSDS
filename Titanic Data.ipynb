#1 Overfitting is what happens when your model performs well on the training data but not on the generalize well.  This can happen when there is too much noise in the model or the sample size is too small.  This is a specific problem  when working with machine learning because when the model detects a new pattern, it will be unable to generalize this to a new data set.  We can prevent overfitting by simplfying the model, gathering more train data or removing the errors and outliers in the data set.
